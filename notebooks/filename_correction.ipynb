```python
def transform_names(dataframe, column_name):
    names = dataframe[column_name].str.split(' ')
    last_names = names.str[-1]
    first_letters = names.str[0].str[0]
    dataframe['author_name_short'] = last_names + first_letters
    return dataframe
```

```python
df = df.drop(columns='New Column')
```

```python
def combine_columns(dataframe, new_column, columns_to_combine):
    dataframe[new_column] = dataframe[columns_to_combine].astype(str).apply('_'.join, axis=1)
    return dataframe
```

```python
df = combine_columns(df, 'combined_column', ['author_name_short', 'citing_year', 'citing_title', 'citing_publication_info'])
```
